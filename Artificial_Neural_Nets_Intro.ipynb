```python
#Perceptron

import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.datasets import load_iris

iris=load_iris()
x=iris.data[:,(2,3)]
y=(iris.target==0).astype(int)

percep_model=Perceptron()
percep_model.fit(x,y)

print("predicted value:",percep_model.predict([x[1]]),"actual value:",y[1])
```

```python
#Importing MNIST data

from tensorflow.examples.tutorials.mnist import input_data

mnist=input_data.read_data_sets("/tmp/data")
```

```python
x_train=mnist.train.images
y_train=mnist.train.labels.astype(int)
x_test=mnist.test.images
y_test=mnist.test.labels.astype(int)
```

```python
import tensorflow as tf

config=tf.contrib.learn.RunConfig(tf_random_seed=42)

feature_columns=tf.contrib.learn.infer_real_valued_columns_from_input(x_train)
dnn_clf=tf.contrib.learn.DNNClassifier(hidden_units=[300,100],n_classes=10,feature_columns=feature_columns,config=config)
dnn_clf=tf.contrib.learn.SKCompat(dnn_clf)

dnn_clf.fit(x_train,y_train,batch_size=50,steps=40000)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import log_loss

y_pred=dnn_clf.predict(x_test)
print("accuracy_score",accuracy_score(y_test,y_pred["classes"]))
print("log loss:",log_loss(y_test,y_pred["probabilities"]))
```

```python
#Neural Network in Tensorflow without libraries

tf.reset_default_graph()

from datetime import datetime

timestamp=datetime.utcnow().strftime("%Y%m%d%H%M%S")
logdir="tf_logs/run-{}/".format(timestamp)

n_inputs=28*28
n_hidden1=400
n_hidden2=180
n_hidden3=80
n_outputs=10

n_epochs=400
batch_size=50

x=tf.placeholder(tf.float32,shape=(None,n_inputs),name="X")
y=tf.placeholder(tf.int64,shape=(None),name="Y")

def neuron_layer(X,n_neurons,name,activation=None,bias=True):
    n_input=int(X.get_shape()[1])
    stddev=2/np.sqrt(n_input)
    init=tf.truncated_normal((n_input,n_neurons),stddev=stddev)
    weights=tf.Variable(init,name="weights")
    if bias:
        bias=tf.Variable(tf.zeros([n_neurons]),name="bias")
        z=tf.matmul(X,weights)+bias
    else:
        z=tf.matmul(X,weights)
    if activation is None:
        return z
    else:
        return activation(z)

#hidden1=neuron_layer(x,n_hidden1,name="hidden1",activation=tf.nn.relu)
#hidden2=neuron_layer(hidden1,n_hidden2,"name=hidden2",activation=tf.nn.relu)
#hidden3=neuron_layer(hidden2,n_outputs,name="output",bias=False)

#For fully connected layers tensorflow provides function to provide fully connected layer 
#with default activation as relu 

with tf.name_scope("dnn") as scope:
    hidden1=tf.layers.dense(x,n_hidden1,name="hidden1",activation=tf.nn.relu)
    hidden2=tf.layers.dense(hidden1,n_hidden2,name="hidden2",activation=tf.nn.relu)
    hidden3=tf.layers.dense(hidden2,n_hidden3,name="hidden3",activation=tf.nn.relu)
    logits=tf.layers.dense(hidden3,n_outputs,name="output")

with tf.name_scope("loss") as scope:
    xentropy=tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y,logits=logits)
    loss=tf.reduce_mean(xentropy,name="loss")
    
with tf.name_scope("train") as scope:    
    optimizer=tf.train.GradientDescentOptimizer(learning_rate=0.01)
    training_op=optimizer.minimize(loss)

with tf.name_scope("eval") as scope:
    correct=tf.nn.in_top_k(logits,y,1)
    accuracy=tf.reduce_mean(tf.cast(correct,tf.float32))
    
init=tf.global_variables_initializer()
saver=tf.train.Saver()

file_writer=tf.summary.FileWriter(logdir,tf.get_default_graph())
accuracy_summary=tf.summary.scalar("Accuracy",accuracy)

with tf.Session() as sess:
    init.run()
    for epoch in range(n_epochs):
        for i in range(int(mnist.train.num_examples/batch_size)):
            x_batch,y_batch=mnist.train.next_batch(batch_size)
            training_op.run(feed_dict={x:x_batch,y:y_batch})
        if epoch%40==0:
            test_accuracy_val=accuracy_summary.eval(feed_dict={x:x_test,y:y_test})
            train_accuracy_val=accuracy_summary.eval(feed_dict={x:x_train,y:y_train})
            file_writer.add_summary(test_accuracy_val,epoch)
            file_writer.add_summary(train_accuracy_val,epoch)
            print(epoch,": test:",accuracy.eval(feed_dict={x:x_test,y:y_test})," train:",accuracy.eval(feed_dict={x:x_train,y:y_train}))
            saver.save(sess,"./my_neural.ckpt")  
```
