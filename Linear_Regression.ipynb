```python
import pandas as pd
import numpy
import sklearn.linear_model as lm
import seaborn as sns
import matplotlib.pyplot as plt

```

# Linear Regression

```python
training_set=pd.read_csv("Brain_Body_noheader.txt",sep=",",header=None)
training_set.columns=['X','Y']
```

```python
linear_regressor=lm.LinearRegression()
linear_regressor=linear_regressor.fit(training_set[["X"]],training_set["Y"])
predictions=linear_regressor.predict(training_set[["X"]])
```

## Visualize the Results

```python
sns.set_style("darkgrid")
plt.scatter(training_set["X"],training_set["Y"])
plt.plot(training_set["X"],predictions)
plt.xlabel("X")
plt.xlabel("Y")
plt.show()
```
